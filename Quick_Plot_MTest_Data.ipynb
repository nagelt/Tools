```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Some plot settings
plt.style.use('seaborn-deep')
plt.rcParams['lines.linewidth']= 2.0
plt.rcParams['lines.color']= 'black'
plt.rcParams['legend.frameon']=True
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.family'] = 'serif'
plt.rcParams['legend.fontsize']=16
plt.rcParams['font.size'] = 16
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.left'] = True
plt.rcParams['axes.spines.bottom'] = True
plt.rcParams['axes.axisbelow'] = True
```

```python
def get_header(filename):
    with open(filename) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=' ')
        line_count = 0
        column_names = []
        for row in csv_reader:
            if row[0] == '#':
                s = " ".join(row[3:])
                column_names.append(s)
    return column_names
```

```python
def get_data(filename):
    header = get_header(filename)
    data = pd.read_csv(filename,sep=' ',header=None, comment='#')
    return header, data
```

```python
def plot_all(filename):
    mtest_names, mtest_data = get_data(filename)
    fig, ax = plt.subplots(ncols=1, nrows = len(mtest_names),figsize=(12,6*len(mtest_names)))
    for i in range(len(mtest_names)-1):
        ax[i].plot(mtest_data[0],mtest_data[i+1])
        ax[i].set_xlabel('time')
        ax[i].set_ylabel(mtest_names[i+1])
    fig.tight_layout()
```

```python
def plot_y_vs_x(filename,x,y):
    mtest_names, mtest_data = get_data(filename)
    fig, ax = plt.subplots()
    ax.plot(mtest_data[x],mtest_data[y])
    ax.set_xlabel(mtest_names[x])
    ax.set_ylabel(mtest_names[y])
    fig.tight_layout()
```

```python
def whatsthere(filename):
    content = get_header(filename)
    for i, c in enumerate(content):
        print(i, ": ", c)
    
```
